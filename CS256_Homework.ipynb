```python
# arXivAPI.py
# uses arXiv Python wrapper, much easier to use
!pip install arxiv
import arxiv

client = arxiv.Client()

# Search for the 10 most recent articles matching the keyword "quantum."
search = arxiv.Search(query = "AI", max_results = 50)
prev = []

results = list(client.results(search))
search = arxiv.Search(query = "machine learning", max_results = 50)
resultML = list(client.results(search))

for r in results:
  print("Link:")
  print(r.entry_id)
  print("\nTitle:")
  print(r.title)
  s = ''
  for i in r.authors:
    s += str(i) + ', '
  print("\nAuthors:")
  print(s[:-2])
  print("\nSummary:")
  print(r.summary)
  print("\n----------------------------------------------------------------------------------\n")

  url = r.entry_id
  title = r.title
  authors = s[:-2]
  summary = r.summary
  if len(summary) > 200:
        summary=summary[:196] + "..."

  if title not in prev:
    cursor.execute("INSERT INTO arxiv (url, name, author, description) VALUES (?, ?, ?, ?)", (url, title, authors, summary))
    prev.append(title)
  else:
    print("Duplicate")

for r in resultML:
  print("Link:")
  print(r.entry_id)
  print("\nTitle:")
  print(r.title)
  s = ''
  for i in r.authors:
    s += str(i) + ', '
  print("\nAuthors:")
  print(s[:-2])
  print("\nSummary:")
  print(r.summary)
  print("\n----------------------------------------------------------------------------------\n")

  url = r.entry_id
  title = r.title
  authors = s[:-2]
  summary = r.summary
  if len(summary) > 200:
        summary=summary[:196] + "..."

  if title not in prev:
    cursor.execute("INSERT INTO arxiv (url, name, author, description) VALUES (?, ?, ?, ?)", (url, title, authors, summary))
    prev.append(title)
  else:
    print("Duplicate")
```

```python
# Coursera
!pip install beautifulsoup4
from bs4 import BeautifulSoup
import requests
import json
import re
url = 'https://www.coursera.org/search?query=AI'

try:
    response = requests.get(url=url)
    r = BeautifulSoup(response.text)
    a = r.find_all('div', class_='cds-CommonCard-clickArea')
    for i in a:

      for k in i.find_all('img'):
        try:
          imagePath = re.findall('src="(.*) srcset=', str(k))[0]
          print(imagePath)
        except:
          imagePath = re.findall('src="(.*)/>', str(k))[0]
          print(imagePath)

      for k in i.find_all('div', class_='cds-CommonCard-bodyContent'):
        skills = re.findall("Skills you'll gain<!-- -->: <\/b>(.*)<\/p><\/div><\/div><div class=\"cds-ProductCard-footer\">", str(i))[0]
        print(skills)

      for j in i.find_all('a'):
        title = re.findall('aria-label="(.*) by \D+, \d', str(j))[0]
        author = re.findall('by (.*), \d.\d stars', str(j))[0]
        stars = re.findall(', (\d.\d) stars', str(j))[0]
        url = re.findall(' href=(.*) id', str(j))[0]
        url = 'https://www.coursera.org' + url[1:-1]
        title = title.replace("&amp;", "&")
        print(url)
        print(title)
        print(author)
        print(stars)
        print()

        cursor.execute("INSERT INTO coursera (image, url, name, author, skills, stars) VALUES (?, ?, ?, ?, ?, ?)", (imagePath, url, title, author, skills, stars))

except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")

except ValueError as e:
    print(f"Decoding JSON failed: {e}")
```

```python
# Towards Data Science Blog
#!pip install beautifulsoup4
from bs4 import BeautifulSoup
import requests
import json
import re
url = 'https://towardsdatascience.com/?s=AI'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

try:
    response = requests.get(url=url, headers=headers)
    r = BeautifulSoup(response.text)
    titles = re.findall('target="_self">(.*)</a>', str(r))
    images = re.findall('src="(.*(\.png|\.jpg|.webp))"', str(r))
    links = re.findall('a href="(.*)" target', str(r))
    images = [i[0] for i in images]
    authors = [titles[i] for i in range(len(titles)) if i % 2 == 1]
    titles = [titles[i] for i in range(len(titles)) if i % 2 == 0]

    for l,i,t,a in zip(links, images, titles, authors):
      print(l)
      print(i)
      print(t)
      print(a)
      print()

      cursor.execute("INSERT INTO blogs (url, image, title, author) VALUES (?, ?, ?, ?)", (l, i, t, a))

except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")

except ValueError as e:
    print(f"Decoding JSON failed: {e}")
```

```python
from bs4 import BeautifulSoup
import requests
import re
url = 'https://www.fast.ai/'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

try:
    response = requests.get(url=url, headers=headers)
    r = BeautifulSoup(response.text)
    for i in r.find_all('div', class_='quarto-post image-right'):
      url = "fast.ai" + re.findall('<a class="no-external" href="(.+)">', str(i))[0][1:]
      author = re.findall('<div class="listing-author">\n(.+)\n<\/div>', str(i))[0]
      title = re.findall('<a class="no-external" href=".+">(.+)</a>', str(i))[0]
      img = ""
      try:
        img = "fast.ai" + re.findall('loading="lazy" src="(.+)"\/>', str(i))[0][1:]
      except:
        try:
          img = "fast.ai" + re.findall('card-img" src="(.+)"\/><\/p>', str(i))[0][1:]
        except:
          img = None

      try:
        summary = re.findall('<p>(.+)<\/p>', str(i))[0]
      except:
        summary = re.findall('.html">(.+)<\/a>', str(i))[0]
      title = title.replace("&amp;", "&")

      if "Doing Data Science for Social Good, Responsibly" == title:
        break

      print(img)
      print(url)
      print(title)
      print(author)
      if len(summary) > 200:
        summary=summary[:196] + "..."
      print(summary)
      print()

      cursor.execute("INSERT INTO fastai (url, image, title, author, summary) VALUES (?, ?, ?, ?, ?)", (url, img, title, author, summary))


except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")

except ValueError as e:
    print(f"Decoding JSON failed: {e}")
```
